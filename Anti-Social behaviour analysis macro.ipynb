```python
from scipy.stats import f_oneway  # One way anova test
F_stat, p_val = f_oneway(*groups)
print(f"ANOVA Fâ€‘statistic: {F_stat:.3f}")
print(f"ANOVA pâ€‘value:     {p_val:.3e}")
```

```python
month_labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'] # violin plot for distribution of incidents between months

# 3) Plot
plt.figure(figsize=(12, 6))
plt.violinplot(
    groups,
    positions=range(1, 13),
    showmeans=True,     # draw the mean marker
    showmedians=False,  # medians are inside the violin
    widths=0.7
)
plt.xticks(range(1, 13), month_labels, rotation=45)
plt.xlabel('Month')
plt.ylabel('Daily ASB Incident Count')
plt.title('Distribution of Daily ASB Incidents by Month (Violin Plot)- Centre line of each distribution is the mean')
plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd  # Tukey test for comparing means between each month

# daily is your DataFrame with columns 'DailyCount' and 'Month' (1â€“12)
tukey = pairwise_tukeyhsd(
    endog=daily['count'],
    groups=daily['Month'],
    alpha=0.05
)
tukey.summary()
```

```python
rows = tukey.summary()
header, data_rows = rows[0], rows[1:]
tukey_data=pd.DataFrame(data_rows, columns=header)
tukey_data.to_excel('tukey.xlsx', index=False)
files.download('tukey.xlsx')
```
